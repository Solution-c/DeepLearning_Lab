```python
import numpy as np
import matplotlib.pyplot as plt
import mpl_toolkits as mplot3d

# 자료입력
x1 = np.array([2, 4, 6, 8])
x2 = np.array([0, 4, 2, 3])
y = np.array([81, 93, 91, 97])
z1 = np.array(range(0,10))
z2 = np.array(range(0,10))

# 초기값 설정
a1 = 2 ; a2 = 2 ; b = 79

#학습률과 반복횟수
lr = 0.021
epoch = 2000

#학습시작
for i in range(epoch):
    result = a1*x1+a2*x2+b
    error = y-result
    a1_diff = (-2)*np.mean(error*x1)
    a2_diff = (-2)*np.mean(error*x2)
    b_diff = (-2)*np.mean(error)
    a1 = a1 - lr*a1_diff
    a2 = a2 - lr*a2_diff
    b = b - lr*b_diff
    i += i

print("y = ", a1,"x1+",a2,"x2+",b)

#그래프 그리기
ax = plt.axes(projection = '3d')
ax.set_xlabel('study_hours')
ax.set_ylabel('private_class')
ax.set_zlabel('Score')
ax.dist = 12
ax.scatter(x1, x2, y, color ='r')
z1, z2 = np.meshgrid(z1, z2)
ax.plot_surface(z1, z2, a1*z1+a2*z2+b, alpha = 0.2)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import mpl_toolkits as mplot3d
from operator import matmul

# 자료입력
data = np.array([[2,4,6,8], [0,4,2,3], [81,93,91,97]])
input = data[:len(data)-1]
z1 = np.array(range(0,10))
z2 = np.array(range(0,10))

# 초기값 설정
coefficients = np.array([2, 2, 79])

#학습률과 반복횟수
lr = 0.021
epoch = 2000

#학습시작
for i in range(epoch):
    result = matmul(coefficients[:len(coefficients)-1],input)+coefficients[len(coefficients)-1]
    error = data[2]-result
    diff = []
    for j in range(len(input)):
        j_diff = (-2)*np.mean(error*input[j])
        diff.append(j_diff)
    b_diff = (-2)*np.mean(error)
    diff.append(b_diff)
    diff = np.array(diff)
    coefficients = coefficients - lr*diff
    i += i

print("y = ", coefficients[0],"x1+",coefficients[1],"x2+",coefficients[2])
```
