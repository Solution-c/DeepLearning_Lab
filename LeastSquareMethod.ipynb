```python
import numpy as np
import numpy.linalg as lin
import matplotlib.pyplot as plt

x = [2, 4, 6, 8]
y = [81, 93, 91, 97]
z = np.array(range(10))

mx = np.mean(x)
my = np.mean(y)

divisor = sum([(i-mx)**2 for i in x])
dividened = 0
for i in range(len(x)):
    dividened += (x[i]-mx)*(y[i]-my)

a = dividened/divisor
b = my -(mx*a)

print(mx, my, divisor, dividened, a, b)

plt.scatter(x,y)
plt.plot(z,a*z+b)
plt.show()
```

```python
from operator import matmul


A = [x,[1 for i in range(len(x))]]
inv_A = lin.pinv(A)

X = matmul(y,inv_A)
X
```

```python
fake_a_b = [3, 76]

y_hat = matmul(np.array(fake_a_b), np.array(A))

def mse(array, array_hat):
    return np.mean([(array[i]-array_hat[i])**2 for i in range(len(array))])

for i in range(len(y)):
    print("공부시간:",x[i], "실제점수:", y[i], "예상점수:", y_hat[i])

mse(y,y_hat)

plt.scatter(x,y)
plt.plot(z,3*z+76)
plt.show
```
