```python
from operator import matmul
from pydoc import ErrorDuringImport
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

#
X = np.array([2, 4, 6, 8])
Y = np.array([81, 93, 91, 97])
x = np.array(range(0,10))

#
a = 0
b = 0

#
lr = 0.032


#
epochs = 2001

result = a*X+b
error = Y-result
a_diff = (-2)*np.mean(error*X)

plt.scatter(X,Y)
for i in range(epochs):
    result = a*X+b
    error = Y-result
    mse = np.mean(error**2)
    a_diff = (-2)*np.mean(error*X)
    b_diff = (-2)*np.mean(error)
    a = a - lr*a_diff
    b = b - lr*b_diff
    print(mse)
    plt.plot(x,a*x+b)
    
    i += i
plt.scatter(X,Y)
plt.xlim([0, 10])      # X : [xmin, xmax]
plt.ylim([70, 100])     # Y : [ymin, ymax]
plt.show()
```
